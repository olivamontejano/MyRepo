```python
from collections import defaultdict, deque
import heapq
import time

def dfs(graph, start, visited=None):

    if visited is None:
        visited = set()

    visited.add(start)
    traversal = [start]

    for neighbor in graph[start]:
        if neighbor not in visited:
            traversal.extend(dfs(graph, neighbor, visited))

    return traversal

def bfs_shortest_path(graph, start, goal):

    queue = deque([[start]])
    visited = set()

    while queue:
        path = queue.popleft()
        node = path[-1]

        if node in visited:
            continue
        visited.add(node)

        if node == goal:
            return path

        for neighbor in graph.get(node, []):
            new_path = list(path)
            new_path.append(neighbor)
            queue.append(new_path)

    return None

def a_star_search(graph, start, goal, heuristic):

    open_set = []
    heapq.heappush(open_set, (0 + heuristic[start], 0, [start]))

    visited = set()

    while open_set:
        f_score, g_score, path = heapq.heappop(open_set)
        current = path[-1]

        if current in visited:
            continue
        visited.add(current)


        if current == goal:
            return path


        for neighbor, cost in graph.get(current, {}).items():
            if neighbor not in visited:
                new_g_score = g_score + cost
                new_f_score = new_g_score + heuristic[neighbor]
                heapq.heappush(open_set, (new_f_score, new_g_score, path + [neighbor]))

    return None

def compare_algorithms(graph, heuristic, start, goal):

    start_time = time.time()
    bfs_path = bfs_shortest_path(graph, start, goal)
    bfs_time = time.time() - start_time

    start_time = time.time()
    a_star_path = a_star_search(graph, start, goal, heuristic)
    a_star_time = time.time() - start_time

    print("Comparison of BFS and A* Performance:")
    print(f"BFS Path: {bfs_path}, Time Taken: {bfs_time:.6f} seconds")
    print(f"A* Path: {a_star_path}, Time Taken: {a_star_time:.6f} seconds")


graph = {
    'A': {'B': 1, 'C': 1},
    'B': {'D': 1},
    'C': {'D': 1, 'E': 1},
    'D': {},
    'E': {'F': 1, 'G': 1},
    'F': {},
    'G': {}
}


heuristic = {
    'A': 4,
    'B': 3,
    'C': 3,
    'D': 2,
    'E': 2,
    'F': 0,
    'G': 1
}


start_node = 'A'
goal_node = 'F'


compare_algorithms(graph, heuristic, start_node, goal_node)


print(f"DFS traversal starting from {start_node}: {dfs(graph, start_node)}")

```
